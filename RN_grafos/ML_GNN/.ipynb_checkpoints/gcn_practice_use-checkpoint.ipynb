# Ejemplo GCN y Node2vec: diferenciar nodos

## Grafo de ejemplo

```python
import networkx as nx

G = nx.from_edgelist([
    (1, 2),
    (1, 5),
    (2, 5),
    (5, 4),
    (4, 3)
])

node_colors = {
    1: "royalblue",
    2: "darkorange",
    3: "forestgreen",
    4: "crimson",
    5: "purple"
}

nx.draw_networkx(G, with_labels=True, node_color=[node_colors[n] for n in G.nodes])
```

## Embeddings Node2vec

```python
import torch
from torch_geometric.nn import Node2Vec
from torch_geometric.utils import from_networkx

data = from_networkx(G)

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = Node2Vec(
    data.edge_index,
    embedding_dim=2,
    walk_length=20,
    context_size=10,
    walks_per_node=10,
    num_negative_samples=1,
    p=1.0,
    q=1.0,
    sparse=True,
).to(device)

loader = model.loader(batch_size=128, shuffle=True, num_workers=4)
optimizer = torch.optim.SparseAdam(model.parameters(), lr=0.01)

def train(model, loader, optimizer):
    model.train()
    total_loss = 0
    for pos_rw, neg_rw in loader:
        optimizer.zero_grad()
        loss = model.loss(pos_rw.to(device), neg_rw.to(device))
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

for epoch in range(1, 201):
    loss = train(model, loader, optimizer)
    if epoch % 10 == 0:
        print(f"Iterations: {epoch}, Loss:{loss:.4f}")
```

```python
import seaborn as sns

def draw_node2vec_embeddings(model, ax):
    model.eval()
    Z = model().clone().detach().cpu().numpy()
    
    axs = sns.scatterplot(x=Z[:,0], y=Z[:,1], ax=ax, hue=node_colors.values(), s=50, palette="tab10")
    
    # put the text of each node near the corresponding node
    for i in range(G.number_of_nodes()):
        axs.text(Z[i,0] + 0.01, Z[i,1], i, fontsize=12)

    ax.set_title("Resultados Node2vec")
    ax.get_legend().remove()

import matplotlib.pyplot as plt

sns.set_style("darkgrid")

fig, ax = plt.subplots()
draw_node2vec_embeddings(model, ax)
plt.show()
```

## Embeddings GCN

```python
from torch.nn import Module
from torch_geometric.nn import GCN, MLP

class GCNMLP(Module):

    def __init__(self, num_features, num_classes, hidden_channels=32, n_layers=2):
        super(GCNMLP, self).__init__()
        self.gcn = GCN(
            in_channels=num_features,
            hidden_channels=hidden_channels,
            num_layers=n_layers,
            out_channels=hidden_channels,   
        )
        self.cls = MLP(
            in_channels=hidden_channels, 
            hidden_channels=hidden_channels,
            out_channels=num_classes, 
            num_layers=2,
            dropout=0.5
        )
    
    def reset_parameters(self):
        self.gcn.reset_parameters()
        self.cls.reset_parameters()

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        x = self.gcn(x, edge_index)
        x = self.cls(x)
        return x
```

```python
from torch.nn import CrossEntropyLoss
from torch.optim import Adam
from torch_geometric.utils import from_networkx
from sklearn.preprocessing import OneHotEncoder

nx.set_node_attributes(G, [1.0], "x")

encoder = OneHotEncoder(categories=[sorted(list(G.nodes))])
nx.set_node_attributes(G, {n:encoder.fit_transform([[n]]).toarray()[0,:] for n in G.nodes}, "y")
data = from_networkx(G)

model = GCNMLP(num_features=1, num_classes=5, hidden_channels=2, n_layers=2)
model.reset_parameters()

optm_func = CrossEntropyLoss()
optimizer = Adam(model.parameters(), lr=0.00001, weight_decay=5e-4)

model.train()
for epoch in range(1001):
    optimizer.zero_grad()
    out = model(data)
    loss = optm_func(out, data.y)
    loss.backward()
    optimizer.step()
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss {loss.item()}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

model.eval()
Z = model.gcn(data.x, data.edge_index).detach().cpu().numpy()

sns.set_style("darkgrid")
f, ax = plt.subplots()

ax = sns.scatterplot(x=Z[:,0], y=Z[:,1], ax=ax, hue=[node_colors[n] for n in G.nodes], s=50, palette="tab10")

for i in range(G.number_of_nodes()):
    ax.text(Z[i,0] + 0.001, Z[i,1], i+1, fontsize=12)

ax.set_title("Resultados GCN")
ax.get_legend().remove()
```

# Neighbor Loader
Es una herramienta para muestrear grafos en mini-lotes (mini-batches) de manera eficiente. Es particularmente útil cuando se trabaja con grafos muy grandes que no caben en memoria o cuando se quiere entrenar una GNN utilizando estrategias de muestreo de vecinos (neighbor sampling).

NeighborLoader crea mini-lotes dinámicos durante el entrenamiento de una GNN, seleccionando un subconjunto de nodos y sus vecinos de manera recursiva. Esto reduce la cantidad de datos procesados en cada iteración, manteniendo la conectividad necesaria para entrenar el modelo.

NeighborLoader se alinea más con el enfoque de GraphSAGE, ya que implementa el muestreo local de vecinos

**diferenciar entre disjoint y joint**
* Implementación disponible en las clase [NeighborLoader](https://pytorch-geometric.readthedocs.io/en/2.5.1/modules/loader.html#torch_geometric.loader.NeighborLoader).

## ¿ Qué hace ?

### Creamos grafo de ejemplo

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.from_edgelist([
    (0, 1), 
    (0, 2), 
    (0, 3), 
    (1, 2), 
    (2, 4), 
    (2, 5), 
    (4, 5)
])

nx.set_node_attributes(G, {n: [n,n] for n in G.nodes()}, 'x')

plt.figure(figsize=(3, 3))
colors = {
    0: "yellow",
    1: "red",
    2: "green",
    3: "royalblue",
    4: "magenta",
    5: "purple"
}
labels = {
    0: "A",
    1: "B",
    2: "C",
    3: "D",
    4: "E",
    5: "F"
}
nx.draw(G, with_labels=True, node_color=[colors[i] for i in range(6)], labels=labels)
```

### Qué hace el loader

#### Disjoint

```python
def draw_batch_disjoint(sub_data, ax, title):
    node_ids = sub_data.batch*10  + sub_data.n_id
    original_edges = node_ids[sub_data.edge_index]
    G_batch = nx.from_edgelist(original_edges.T.numpy())

    node_colors = sub_data.batch.tolist()
    node_labels = {n:labels[n%10] for n in G_batch.nodes()}
    edge_colors = ["red"]*sub_data.batch_size + ["white"]*(len(sub_data.n_id) - sub_data.batch_size)

    pos = nx.spring_layout(G_batch, k=0.75)

    nx.draw_networkx(
        G_batch, 
        ax=ax,
        pos=pos,
        with_labels=True, 
        node_color=node_colors, 
        labels=node_labels,
        edgecolors=edge_colors,
        linewidths=2,
        nodelist=node_ids.tolist(),
        cmap="tab10"
    )
    ax.set_title(title)
```

```python
# Importar la función from_networkx para convertir un grafo de NetworkX a un objeto compatible con PyTorch Geometric
from torch_geometric.utils import from_networkx  
# Importar NeighborLoader para muestreo eficiente de nodos y sus vecinos en grafos grandes
from torch_geometric.loader import NeighborLoader  

# Convertir un grafo de NetworkX (G) en un objeto de datos de PyTorch Geometric
data = from_networkx(G)  

# Definir el número de vecinos a muestrear en cada capa de la GNN
# En este caso: 2 vecinos para la primera capa, 1 vecino para la segunda capa
neis = [2, 1]  

# Crear un NeighborLoader para muestrear mini-lotes de nodos y sus vecinos
train_loader = NeighborLoader(
    data,               # El objeto de datos que representa el grafo
    batch_size=3,       # Tamaño del batch inicial (3 nodos de partida)
    num_neighbors=neis, # Número de vecinos a muestrear en cada capa
    shuffle=True,       # Mezclar los nodos iniciales en cada iteración
    disjoint=True       # Garantizar que los mini-lotes sean disjuntos (sin solapamiento)
)  

# Iterar sobre los mini-lotes generados por el NeighborLoader
for batch in train_loader:  
    # Crear una figura y un eje para graficar
    fig, axs = plt.subplots()  
    # Dibujar el mini-lote actual como un subgrafo
    # `draw_batch_disjoint` es una función personalizada que visualiza el subgrafo disjunto
    draw_batch_disjoint(batch, axs, "Batch")  
```

#### Joint

```python
def draw_batch(sub_data, ax, title):
    original_edges = sub_data.n_id[sub_data.edge_index]
    G_batch = nx.from_edgelist(original_edges.T.numpy())
    plt.figure(figsize=(3, 3))

    node_colors = [colors[n.item()] for n in sub_data.n_id]
    node_labels = {n:labels[n] for n in G_batch.nodes()}
    edge_colors = ["black"]*sub_data.batch_size + ["white"]*(len(sub_data.n_id) - sub_data.batch_size)

    nx.draw(
        G_batch, 
        ax=ax,
        with_labels=True, 
        node_color=node_colors, 
        labels=node_labels,
        edgecolors=edge_colors,
        nodelist=sub_data.n_id.tolist(),
        linewidths=2,
    )
    ax.set_title(title)
```

```python
from torch_geometric.utils import from_networkx
from torch_geometric.loader import NeighborLoader

data = from_networkx(G)

neis = [2, 1]
train_loader = NeighborLoader(data, batch_size=2, num_neighbors=neis, shuffle=True, disjoint=False)

N_batches = len(train_loader)
fig, axs = plt.subplots(N_batches, figsize=(3, 3*N_batches))
for i, sub_data in enumerate(train_loader):
    draw_batch(sub_data, axs[i], f"Batch {i}")
```

## ¿Cómo lo uso?

### Load Data

```python
from torch_geometric.datasets import CitationFull

dataset = CitationFull(root='CitationFull', name='Cora_ML')
print(f"Number of graphs: {len(dataset)}")

data = dataset[0]
```

### Random split

```python
import torch
import numpy as np
from sklearn.model_selection import train_test_split

train_idx, test_idx = train_test_split(np.arange(data.num_nodes), test_size=0.2, stratify=data.y)

train_mask = np.zeros(data.num_nodes, dtype=bool)
train_mask[train_idx] = True

test_mask = np.zeros(data.num_nodes, dtype=bool)
test_mask[test_idx] = True

data.train_mask = torch.tensor(train_mask)
data.test_mask = torch.tensor(test_mask)
```

### Creo red GIN

```python
from torch.nn import Module
from torch_geometric.nn import GIN, MLP

class GINMLP(Module):

    def __init__(self, num_features, num_classes, hidden_channels=32, n_layers=2):
        super(GINMLP, self).__init__()
        self.gcn = GIN(
            in_channels=num_features,
            hidden_channels=hidden_channels,
            num_layers=n_layers,
            out_channels=hidden_channels,
            dropout=0.5,
            act="relu",
            jk="cat"   
        )
        self.cls = MLP(
            in_channels=hidden_channels, 
            hidden_channels=hidden_channels,
            out_channels=num_classes, 
            num_layers=2,
            dropout=0.5
        )
    
    def reset_parameters(self):
        self.gcn.reset_parameters()
        self.cls.reset_parameters()

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        x = self.gcn(x, edge_index)
        x = self.cls(x)
        return x
```

### Train

```python
from torch.nn import CrossEntropyLoss
from torch.optim import Adam
from torch_geometric.loader import NeighborLoader

def train():
    model.train()
    
    loss_all = 0
    for batch in train_loader:
        batch = batch.to(device)
        batch_size = batch.batch_size
        
        optimizer.zero_grad()
        out = model(batch)
        loss = optm_func(out[:batch_size], batch.y[:batch_size])
        loss.backward()
        optimizer.step()

        loss_all += loss.item()
    
    return loss_all/len(train_loader)

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = GINMLP(num_features=data.num_features, num_classes=dataset.num_classes)
model.reset_parameters()
model = model.to(device)

neis = [25, 10]
train_loader = NeighborLoader(data, batch_size=32, num_neighbors=neis, input_nodes=data.train_mask, shuffle=True)

optimizer = Adam(model.parameters(), lr=0.01, weight_decay=5e-4)
optm_func = CrossEntropyLoss()
optm_func = optm_func.to(device)

model.train()
for epoch in range(50):
    
    loss = train()
    
    if epoch % 5 == 0:
        print(f"Epoch {epoch}: Loss {loss}")
```

### Test

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, f1_score

def test():
    model.eval()
    
    y_pred = []
    y_true = []
    with torch.no_grad():
        for data in test_loader:
            data = data.to(device)
            out = model(data)
            y_pred.extend(out.argmax(dim=1)[:data.batch_size].tolist())
            y_true.extend(data.y[:data.batch_size].tolist())
        
    return y_true, y_pred

device = 'cuda' if torch.cuda.is_available() else 'cpu'

test_loader = NeighborLoader(data, batch_size=len(test_idx), num_neighbors=[-1, -1], input_nodes=data.test_mask, shuffle=False)
y, y_pred = test()

print(f"F1 score: {f1_score(y, y_pred, average='micro')}")
sns.heatmap(confusion_matrix(y, y_pred), annot=True, fmt='g', cmap='Blues')
```

### Genero batches juntando comunidades

```python
from torch_geometric.utils import to_networkx

def draw_batch_cluster(data, ax, title):

    g_batch = to_networkx(data, to_undirected=True)
    node_labels = {n:data.id[n].item() for n in g_batch.nodes()}
    node_colors = [node_coms[data.id[n].item()] for n in g_batch.nodes()]
    pos = nx.spring_layout(g_batch, k=0.75)
    nx.draw_networkx(g_batch, pos=pos, with_labels=True, labels=node_labels, node_color=node_colors, ax=ax)
    ax.set_title(title)
```

```python
from torch_geometric.loader import ClusterData, ClusterLoader

cluster_data = ClusterData(data, num_parts=5, keep_inter_cluster_edges=True)
loader = ClusterLoader(cluster_data, batch_size=2)

N_batches = len(loader)
fig, axs = plt.subplots(N_batches, figsize=(4, 4*N_batches))
for i, sub_data in enumerate(loader):
    draw_batch_cluster(sub_data, axs[i], f"Batch {i}")
```

# GraphSAINT

* [GraphSAINTNodeSampler](https://pytorch-geometric.readthedocs.io/en/stable/modules/loader.html#torch_geometric.loader.GraphSAINTNodeSampler): Muestra nodos al azar siendo la probabilidad de elegir cada nodo $ P(u) \propto \left\| \tilde{A}_{:,u} \right\|^2 $
* [GraphSAINTEdgeSampler](https://pytorch-geometric.readthedocs.io/en/stable/modules/loader.html#torch_geometric.loader.GraphSAINTEdgeSampler): Muestre aristas al azar siendo la probabilidad de elegir cada arista $ P(e_{u,v}) \propto \frac{1}{\text{deg}(u)} + \frac{1}{\text{deg}(v)}$
* [GraphSAINTRandomWalkSampler](https://pytorch-geometric.readthedocs.io/en/stable/modules/loader.html#torch_geometric.loader.GraphSAINTRandomWalkSampler): Se realizan $r$ nodos al azar siguiendo una distribución uniforme y luego se realizan caminos aleatorios de longitud $h$ empezando en cada nodo.

## ¿Qué hace?

### Cargar grafo de prueba

```python
import networkx as nx
from torch_geometric.utils import from_networkx

G = nx.karate_club_graph()
nx.set_node_attributes(G, {n: [G.degree(n)] for n in G.nodes()}, "x")
nx.set_node_attributes(G, {n: n for n in G.nodes()}, "id")

for n in G.nodes():
    G.nodes[n].pop("club")

for u, v in G.edges():
    G[u][v].pop("weight")

data = from_networkx(G)

nx.draw_networkx(G, with_labels=True)
```

### Node sampler

* `batch_size`: Nº de nodos elegidos (aprox)
* `num_steps`: Nº de grafos generados

```python
from torch_geometric.utils import to_networkx

def draw_batch_saintnode(batch, ax, title):
    g_batch = to_networkx(batch, to_undirected=True)
    nx.relabel_nodes(g_batch, {n: batch.id[n].item() for n in g_batch.nodes()}, copy=False)
    node_colors = batch.node_norm.tolist()
    edge_colors = batch.edge_norm.tolist()

    pos = nx.spring_layout(g_batch, k=0.5)
    nx.draw_networkx(
        g_batch, 
        pos=pos, 
        with_labels=True, 
        ax=ax, 
        node_color=node_colors,
        cmap="Blues",
        edge_color=edge_colors,
        edge_cmap=plt.cm.plasma,
        nodelist=batch.id.tolist(), 
        edgelist=batch.id[batch.edge_index].T.tolist(), 
        edgecolors="navy"
    )
    ax.set_title(title)
```

```python
import matplotlib.pyplot as plt
from torch_geometric.loader import GraphSAINTNodeSampler

sampler = GraphSAINTNodeSampler(data, num_steps=5, batch_size=16, sample_coverage=100)

N_batches = len(sampler)
fig, axs = plt.subplots(N_batches, figsize=(4, 4*N_batches))

for i,batch in enumerate(sampler):
    draw_batch_saintnode(batch, axs[i], f"Batch {i}")
```

## ¿Cómo lo uso?

### Load Data

```python
from torch_geometric.datasets import CitationFull
from torch_geometric.utils import degree

dataset = CitationFull(root='CitationFull', name='Cora_ML')
print(f"Number of graphs: {len(dataset)}")

data = dataset[0]

# para poder aplicar la normalizacion de alpha que se aplica en GraphSAINT es necesario darle un peso a cada arista
# de primeras cada arista tiene la misma probabilidad de ser seleccionada
row, col = data.edge_index
data.edge_weight = 1. / degree(col, data.num_nodes)[col] 
```

### Random split

```python
import torch
import numpy as np
from sklearn.model_selection import train_test_split

train_idx, test_idx = train_test_split(np.arange(data.num_nodes), test_size=0.2, stratify=data.y)

train_mask = np.zeros(data.num_nodes, dtype=bool)
train_mask[train_idx] = True

test_mask = np.zeros(data.num_nodes, dtype=bool)
test_mask[test_idx] = True

data.train_mask = torch.tensor(train_mask)
data.test_mask = torch.tensor(test_mask)
```

### Creo red GIN

```python
from torch.nn import Module
from torch_geometric.nn import GIN, MLP

class GINMLP(Module):

    def __init__(self, num_features, num_classes, hidden_channels=32, n_layers=2):
        super(GINMLP, self).__init__()
        self.gcn = GIN(
            in_channels=num_features,
            hidden_channels=hidden_channels,
            num_layers=n_layers,
            out_channels=hidden_channels,
            dropout=0.5,
            act="relu",
            jk="cat"   
        )
        self.cls = MLP(
            in_channels=hidden_channels, 
            hidden_channels=hidden_channels,
            out_channels=num_classes, 
            num_layers=2,
            dropout=0.5
        )
    
    def reset_parameters(self):
        self.gcn.reset_parameters()
        self.cls.reset_parameters()

    def forward(self, x, edge_index, edge_weight=None):
        x = self.gcn(x, edge_index, edge_weight)
        x = self.cls(x)
        return x.log_softmax(dim=-1)
```

### Train

```python
import torch.nn.functional as F
from torch.optim import Adam
from torch_geometric.loader import GraphSAINTRandomWalkSampler

def train():
    model.train()
    
    loss_all = 0
    for batch in train_loader:
        batch = batch.to(device)
        
        optimizer.zero_grad()
        # calculamos el nuevo peso de las aristas en base a la normalizacion de alpha
        edge_weight = batch.edge_norm * batch.edge_weight # importancia que se le da a cada arista
        out = model(batch.x, batch.edge_index, edge_weight)
        
        # calculamos el cross entropy loss en dos pasos, primero calculamos la loss para cada nodo
        # luego aplicamos la normalizacion de lambdas y luego sumamos las losses
        loss = F.nll_loss(out, batch.y, reduction='none')
        loss = (loss * batch.node_norm)[batch.train_mask].sum()

        # paso de optimizacion
        loss.backward()
        optimizer.step()

        loss_all += loss.item()
    
    return loss_all/len(train_loader)

device = 'cuda' if torch.cuda.is_available() else 'cpu'

model = GINMLP(num_features=data.num_features, num_classes=dataset.num_classes)
model.reset_parameters()
model = model.to(device)

train_loader = GraphSAINTRandomWalkSampler(data, walk_length=3, num_steps=10, batch_size=32, sample_coverage=100)

optimizer = Adam(model.parameters(), lr=0.01, weight_decay=5e-4)

model.train()
for epoch in range(61):
    
    loss = train()
    
    if epoch % 5 == 0:
        print(f"Epoch {epoch}: Loss {loss}")
```

### Test

# Graph Pooling: Mean

## Definir red

Utilizo la clase de pytorch geometric que implementa una [GCN](https://pytorch-geometric.readthedocs.io/en/latest/generated/torch_geometric.nn.models.GCN.html#torch_geometric.nn.models.GCN)

$ \mathbf{z}_i^{l} = W^l \sum_{j \in \mathcal{N}(i) \cup \{i\}} (\frac{\mathbf{x}_j^{l-1}}{\sqrt{d_j d_i}}) + b^l$

$ \mathbf{z}_G = \frac{1}{|V|} * \sum_{v \in V}(\mathbf{z}_v) $

```python
from torch.nn import Module
from torch_geometric.nn import GCN, MLP, global_mean_pool # CALCULA LA MEDIA DE TODOS LOS EMBEDDINGS DE LOS NODOS

class GCNGraph(Module):

    def __init__(self, num_features, num_classes, hidden_channels=32, n_layers=2):
        super(GCNGraph, self).__init__()
        self.gcn = GCN(
            in_channels=num_features,
            hidden_channels=hidden_channels,
            num_layers=n_layers,
            out_channels=hidden_channels,
            jk="cat",
            dropout=0.5    
        )
        self.cls = MLP(
            in_channels=hidden_channels, 
            hidden_channels=hidden_channels,
            out_channels=num_classes, 
            num_layers=2,
            dropout=0.5
        )
    
    def reset_parameters(self):
        self.gcn.reset_parameters()
        self.cls.reset_parameters()

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        
        x = self.gcn(x, edge_index)
        x = global_mean_pool(x, batch) # EN SUSTITUCION A ESTE MEAN_POOL, SE PUEDEN USAR LOS AGREGADORES DEL OTRO NOTEBOOK
        x = self.cls(x)
        return x
```

# Graph pooling GRU

## GRU pooling Aggregator

$ \mathbf{q}_t = \text{GRU}(\mathbf{o}_{t-1}, \mathbf{q}_{t-1}) $

$ e_{v,t} = \sum \mathbf{z}_v \odot \mathbf{q}_t $

$ a_{v,t} = \frac{\exp(e_{v,t})}{\sum_{u \in \mathcal{V}} \exp(e_{u,t})} $

$ \mathbf{o}_t = \sum_{v \in \mathcal{V}} a_{v,t} \mathbf{z}_v $

$z_G = (o_1 \oplus o_2 \oplus ... \oplus o_T) $

### GRU update



```python
import torch
import torch.nn as nn

class UpdateGRU(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        """
        Custom GRU based on the given equations.
        
        Args:
            input_dim (int): Dimension of the input vector x.
            hidden_dim (int): Dimension of the hidden state h.
        """
        super(UpdateGRU, self).__init__()
        self.hidden_dim = hidden_dim
        
        # Input-to-hidden weights and biases
        self.W_ir = nn.Linear(input_dim, hidden_dim, bias=True)
        self.W_iz = nn.Linear(input_dim, hidden_dim, bias=True)
        self.W_in = nn.Linear(input_dim, hidden_dim, bias=True)
        
        # Hidden-to-hidden weights and biases
        self.W_hr = nn.Linear(hidden_dim, hidden_dim, bias=True)
        self.W_hz = nn.Linear(hidden_dim, hidden_dim, bias=True)
        self.W_hn = nn.Linear(hidden_dim, hidden_dim, bias=True)
        
        # Activation functions
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()

    def reset_parameters(self):
        self.W_ir.reset_parameters()
        self.W_iz.reset_parameters()
        self.W_in.reset_parameters()
        self.W_hr.reset_parameters()
        self.W_hz.reset_parameters()
        self.W_hn.reset_parameters

    def forward(self, x_t, h_t_minus_1):
        """
        Forward pass for a single time step.
        
        Args:
            x_t (torch.Tensor): Input vector at time t, shape (batch_size, input_dim).
            h_t_minus_1 (torch.Tensor): Hidden state at time t-1, shape (batch_size, hidden_dim).
        
        Returns:
            h_t (torch.Tensor): Updated hidden state at time t, shape (batch_size, hidden_dim).
        """
        # Compute reset gate
        r_t = self.sigmoid(self.W_ir(x_t) + self.W_hr(h_t_minus_1))
        
        # Compute update gate
        z_t = self.sigmoid(self.W_iz(x_t) + self.W_hz(h_t_minus_1))
        
        # Compute candidate hidden state
        n_t = self.tanh(self.W_in(x_t) + r_t * self.W_hn(h_t_minus_1))
        
        # Compute new hidden state
        h_t = (1 - z_t) * n_t + z_t * h_t_minus_1
        
        return h_t
```

### Aggregator class

```python
import torch.nn.functional as F
from torch.nn import Linear
from torch_geometric.nn import Aggregation
from torch_geometric.utils import softmax, scatter

class GlobalGRUPooling(Aggregation):

    def __init__(self, input_dim, T):
        super(GlobalGRUPooling, self).__init__()
        self.gru = UpdateGRU(input_dim, input_dim)
        self.T = T
        self.input_dim = input_dim

    def reset_parameters(self):
        self.gru.reset_parameters()
        self.mlp_alpha.reset_parameters()

    def forward(self, x, index, ptr=None, dim_size=None, dim=-2, max_num_elements=None):
        batch_size = ptr.shape[0]-1

        q = torch.zeros(batch_size, self.input_dim)
        o = torch.zeros(batch_size, self.input_dim)

        outputs = []

        for _ in range(self.T): # T es un hiperparametro
            q = self.gru(o, q)
            
            q_hat = q[index]
            e =  torch.sum(x * q_hat, dim=1, keepdim=True)
            a = softmax(e, index)

            o = scatter(a*x, index, reduce='sum') # aquellas entradas a*x que tengan el mismo indice "index" se les aplica la suma
            outputs.append(o)

        return torch.cat(outputs, dim=1)
```

## Definir GNN

$ \mathbf{z}_i^{l} = W^l \sum_{j \in \mathcal{N}(i) \cup \{i\}} (\frac{\mathbf{x}_j^{l-1}}{\sqrt{d_j d_i}}) + b^l$

$ z_G = (o_1 \oplus o_2 \oplus ... \oplus o_T) $

```python
from torch.nn import Module
from torch_geometric.nn import GCN, MLP

class GCNGlobalGRUPooling(Module):

    def __init__(self, num_features, num_classes, hidden_channels=64, n_layers=3, T=5):
        super(GCNGlobalGRUPooling, self).__init__()
        self.gcn = GCN(
            in_channels=num_features,
            hidden_channels=hidden_channels,
            num_layers=n_layers,
            out_channels=hidden_channels,
            dropout=0.5,
            jk="cat",    
        )
    
        self.gru_pooling = GlobalGRUPooling(hidden_channels, T)

        self.cls = MLP(
            in_channels=hidden_channels*T, 
            hidden_channels=hidden_channels,
            out_channels=num_classes, 
            num_layers=2,
            dropout=0.5
        )
    
    def reset_parameters(self):
        self.gcn.reset_parameters()
        self.cls.reset_parameters()
        self.gru_pooling.reset_parameters()

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        x = self.gcn(x, edge_index)
        x = self.gru_pooling(x, data.batch, ptr=data.ptr)
        x = self.cls(x)
        return x
```

## Optimizar

```python
from torch.nn.functional import sigmoid
from torch.optim import Adam

gnn.reset_parameters()

def loss_func(data, A, k=5):
    Z = gnn(data.x, data.edge_index) # calculo embeddings
    A_hat = sigmoid(Z @ Z.T) #aplico sigmoid de la funcion
    # Evito que el valor sea 0 o 1 exactamente para evitar problemas con el logaritmo
    A_hat = torch.clamp(A_hat, min=1e-7, max=1 - 1e-7) # 
    loss = -torch.sum(k*(A * torch.log(A_hat)) + (1 - A) * torch.log(1 - A_hat))
    return loss

A = torch.tensor(nx.to_numpy_array(G))

optimizer = Adam(gnn.parameters(), lr=0.005, weight_decay=5e-4)

gnn.train()
for epoch in range(2001):
    optimizer.zero_grad()
    loss = loss_func(data, A, 10)
    loss.backward()
    optimizer.step()
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss {loss.item()}")
```
