```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import networkx as nx
import numpy as np
import pandas as pd
from pyvis.network import Network

def str_color(r,g,b,a):
    return "#" + format(int(r*255), f'0{2}x') + format(int(g*255), f'0{2}x') + format(int(b*255), f'0{2}x') + format(int(a*255), f'0{2}x')

def assign_colors(values, vmin=None, vmax=None, cmap=None):
    colormap = plt.cm.magma if cmap is None else cmap

    # Normalize your values to be in the range [0, 1] if they are not already
    if vmin is None:
        vmin = min(values)
    if vmax is None:
        vmax = max(values)
    norm = mcolors.Normalize(vmin=vmin, vmax=vmax)

    # Create a colormap scalar mappable
    sm = plt.cm.ScalarMappable(cmap=colormap, norm=norm)
    sm.set_array([])

    # Map your values to colors
    colors = [str_color(*colormap(norm(value))) for value in values]
    return colors

def draw_centrality(r, ax, pos=None, vmin=None, vmax=None):
    colors = assign_colors(r, vmin, vmax)
    nx.draw(G, ax=ax, pos=pos, with_labels=True, node_color=colors)

def make_graph():
    G = nx.florentine_families_graph()
    G = G.to_directed()
    G.add_edges_from([
        ('Navio', 'Panizo'),
        ('Panizo', 'Navio'),
        ('Garcia', 'Panizo'),
        ('Gonzalez', 'Panizo'),
        ('Panizo', 'Perez',),
        ('Panizo', 'Montes'),
        ('Panizo', 'Lopez'),
        ('Panizo', 'Lamberteschi'),
        ('Lamberteschi', 'Panizo')
    ])
    return G
```

1. Base Matem√°tica:
Eigenvector Centrality:

Se basa puramente en el c√°lculo de los vectores propios de la matriz de adyacencia del grafo.
Un nodo es importante si est√° conectado a otros nodos importantes. No incluye mecanismos adicionales como ponderaciones externas o factores de aleatoriedad.
PageRank:

Utiliza una matriz de transici√≥n basada en la estructura del grafo, pero tambi√©n incluye un factor de amortiguaci√≥n (damping factor) para simular el comportamiento aleatorio de un usuario que "salta" entre nodos.
Esto introduce la probabilidad de "teletransportarse" a cualquier nodo en el grafo, haciendo que incluso nodos aislados tengan una peque√±a importancia.

2. Influencia de las Conexiones:
Eigenvector Centrality:

Considera que todas las conexiones (aristas) tienen el mismo peso. En grafos no dirigidos, esto significa que la importancia de un nodo solo depende de la importancia de sus vecinos inmediatos.
PageRank:

Da menos peso a nodos con muchas conexiones (diluye su influencia). Por ejemplo, un nodo conectado a muchos otros distribuye su importancia entre todos ellos, mientras que en eigenvector centrality, esto no ocurre.
Introduce un mecanismo probabil√≠stico: un nodo conectado a pocos pero importantes nodos puede tener m√°s peso.

| **Caracter√≠stica**            | **Eigenvector Centrality**                 | **PageRank**                            |
|--------------------------------|--------------------------------------------|-----------------------------------------|
| **C√°lculo**                   | Vectores propios de la matriz de adyacencia | Modelo de transici√≥n + damping factor  |
| **Conexiones**                | Todas las conexiones tienen igual peso      | Considera diluci√≥n en nodos de grado alto |
| **Normalizaci√≥n**             | No necesariamente                         | Normalizado como probabilidad          |
| **Flexibilidad**              | Menos flexible                            | M√°s adaptable (damping factor)         |
| **Aplicaci√≥n t√≠pica**         | Redes sociales, redes biol√≥gicas           | Web ranking, grafos con saltos aleatorios |

### Explicaci√≥n de como funciona
Si hacemos varios paseos aleatorios cual es la probabilidad de visitar cada nodo

```python
G = nx.DiGraph()
G.add_edges_from([
    ("a","a"),
    ("a","b"),
    ("c","b"),
    ("b","a"),
    ("b","c")
])

fig = plt.figure(figsize=(3,3))
nx.draw(G, with_labels=True)
plt.show(fig)
```

#### Simulaci√≥n paseos aleatorios
* Ejecuto suficiente paseos aleatorios y luego cuento el porcentaje de veces que vi cada nodo

```python
from matplotlib.animation import FFMpegWriter
import matplotlib.animation as animation

def do_step(G,node): # camino aleatorio
    return np.random.choice(np.array(list(nx.neighbors(G,node))),1)[0]

def simulate_random_walk(G, steps):
    init = np.random.choice(np.array(list(G.nodes)),1)[0]
    
    visited = [init]
    for _ in range(steps):
        next = do_step(G, init)
        visited.append(next)
        init = next
    
    return visited

def calculate_r(G, visited):
    values, counts = np.unique(visited, return_counts=True)

    missing = list(set(G.nodes) - set(values)) # pone a 0 los no visitados
    values = np.append(values, missing)
    counts = np.append(counts, np.zeros(len(missing)))
    
    r = counts[np.argsort(values)]/np.sum(counts)
    
    return r
```

##### Visualizar pasos

```python
def create_random_walk_video(G, steps, filename="./random_walk.mp4"):
    fig, ax = plt.subplots(figsize=(8, 8))
    pos = nx.kamada_kawai_layout(G)
    visited = []
    
    # Set up video writer
    writer = FFMpegWriter(fps=5, metadata=dict(artist='Me'), bitrate=1800)
    
    with writer.saving(fig, filename, dpi=100):
        for t in range(steps):
            visited.extend(simulate_random_walk(G, 5))
            r = calculate_r(G, visited)
            
            ax.clear()
            ax.set_title(f"Step {t} | r({r[0]:.2f},{r[1]:.2f},{r[2]:.2f})")
            draw_centrality(r, ax, pos, vmin=0, vmax=1)
            writer.grab_frame()
    
    plt.close(fig)
    print(f"Video saved as {filename}")
```

```python
create_random_walk_video(G, 100, "pagerank_rw_simulation.mp4")
```

##### Ejecutar hasta converger

```python
def pagerank_centrality_simulation(G):
    N = 100
    steps = 5
    total = np.array([simulate_random_walk(G,steps) for _ in range(N)]).flatten()
    
    r = calculate_r(G, total)
    
    df_visited = pd.DataFrame(r, index=["a","b","c"], columns=["probability"])
    return df_visited
    
pagerank_centrality_simulation(G)
```

#### Simulacion paseos aleatorios con matrix de adyacencia
* Puedo utilizar la matriz de adyacencia para actualizar la probabilidad de haber visitado cada nodo
    * p(t+1) = M*p(t)

```python
import numpy as np

def init_problem(G):
    N = G.number_of_nodes()
    D = np.repeat(np.array([v for _,v in G.out_degree()]).reshape(1,-1),N, axis=0)
    A = nx.adjacency_matrix(G).todense()
    M = A/D
    
    r = np.zeros(N) + 1/N

    return M, r
```

```python
M, r = init_problem(G)

# La matriz R tiene la prob de visitar cada nodo al principio
print("R:")
display(r)
print("M:")
display(M)
```

##### Ejecutar hasta converger

```python
def pagerank_centrality_matrix(G, M, r):
    
    while True:
        r_next = np.dot(M,np.transpose(r))
    
        error = np.sum(np.abs(r_next - r))
        if error < 1e-5:
            break
    
        r = r_next
    
    df_visited = pd.DataFrame(r, index=["a","b","c"], columns=["probability"])
    return df_visited

M, r = init_problem(G)
pagerank_centrality_matrix(G, M, r)
```

#### Calculo con eigenvectors
* Cuando la simulacion converge entonces:
    * p = M*p
* Esta ecuaci√≥n es muy parecida a la ecuaci√≥n (Œªx = Ax) que usamos para calcular los eigenvectors y los eigenvalues de una matrix (p=x | M=A | Œª=1). 
* Por lo tanto podemos obtener el vector p sin necesidad de simular nada, calculando los eigenvectors y eigenvalues de la matriz M.
* **El vector p es el eigenvector correspondiente al eigenvalue de mayor valor**   

```python
import scipy.sparse as sparse
from scipy.sparse.linalg import eigs

def pagerank_centrality_eigen(M):
    
    # compute the eigenvalues and eigenvectors
    lam, eigenvectors = eigs(M, k=1, which='LR')
    
    # Get the dominant eigenvector
    p = np.real(eigenvectors[:, 0])
    
    # normalize so they add up to 1
    p = p/np.sum(p)
    
    df_visited = pd.DataFrame(p, index=["a","b","c"], columns=["probability"])
    return df_visited

M, _ = init_problem(G)
pagerank_centrality_eigen(M)
```

#### Solucion: Permitir que el random walk se teletransporte
* Permitimos que el caminante aleatorio de teletransporte a otro nodo al azar con una probabilitad baja (sobre 0.1 o 0.2)
* Para ello podemos cambiar la matriz M de la siguiente forma:
    * G = B\*M + (1-B)\*[1/N]NXN
* G tambi√©n es conocida como la matriz Google

```python
B = 0.85
M, r = init_problem(G)

g_matrix = B*M + (1 - B)*np.array([1/3]*(3*3)).reshape(3,3)
g_matrix
```

## Edge Degrees

```python
import networkx as nx
import pandas as pd

def edge_degree(G):

    def _score(s,t):
        return G.degree(s)*G.degree(t)
        
    return {
        (s,t): _score(s,t)
        for s,t in G.edges
    }

G = make_graph()
G = G.to_undirected()

centrality = edge_degree(G)

df = pd.Series(centrality.values(), index=[f"{s}-{t}" for s,t in centrality.keys()], name="score").reset_index()
display(df.sort_values("score", ascending=False).head(5))

nx.set_node_attributes(G, {n:f"{n}" for n in G.nodes}, 'label')
nx.set_node_attributes(G, "dot", 'shape')
nx.set_edge_attributes(G, {n:f"{v:.2f}" for n,v in centrality.items()}, 'title')
nx.set_edge_attributes(G, {n:v for n,v in zip(centrality.keys(), assign_colors(centrality.values(), cmap=plt.cm.Greys))}, 'color')

net = Network(notebook=True, height='500px', width='100%', bgcolor="white")
net.from_nx(G)
net.show('centrality.html')
```

## Topological overlap

```python
import networkx as nx
import pandas as pd

def edge_topological_overlap(G):
    
    def _score(s,t):
        common_neis = len(list(nx.common_neighbors(G,s,t)))
        s_d = G.degree(s)
        t_d = G.degree(t)
        return common_neis/((s_d - 1) + (t_d - 1) + common_neis)

    return {
        (s,t): _score(s,t)
        for s,t in G.edges
    }   

G = make_graph()
G = G.to_undirected()

centrality = edge_topological_overlap(G)

df = pd.Series(centrality.values(), index=[f"{s}-{t}" for s,t in centrality.keys()], name="score").reset_index()
display(df.sort_values("score", ascending=False).head(5))

nx.set_node_attributes(G, {n:f"{n}" for n in G.nodes}, 'label')
nx.set_node_attributes(G, "dot", 'shape')
nx.set_edge_attributes(G, {n:f"{v:.2f}" for n,v in centrality.items()}, 'title')
nx.set_edge_attributes(G, {n:v for n,v in zip(centrality.keys(), assign_colors(centrality.values(), cmap=plt.cm.Greys))}, 'color')

net = Network(notebook=True, height='500px', width='100%', bgcolor="white")
net.from_nx(G)
net.show('centrality.html')
```

## Edge Diffusion of importance

```python
import networkx as nx
import pandas as pd

def edge_difussion_of_importance(G):

    def _not_in_neighbor(s,t):
        return len(set(G.neighbors(t)).difference(set(G.neighbors(s))))
    
    def _score(s,t):
        return (_not_in_neighbor(s,t) + _not_in_neighbor(t,s))/2

    return {
        (s,t): _score(s,t)
        for s,t in G.edges
    }   

G = make_graph()
G = G.to_undirected()

centrality = edge_difussion_of_importance(G)

df = pd.Series(centrality.values(), index=[f"{s}-{t}" for s,t in centrality.keys()], name="score").reset_index()
display(df.sort_values("score", ascending=False).head(5))

nx.set_node_attributes(G, {n:f"{n}" for n in G.nodes}, 'label')
nx.set_node_attributes(G, "dot", 'shape')
nx.set_edge_attributes(G, {n:f"{v:.2f}" for n,v in centrality.items()}, 'title')
nx.set_edge_attributes(G, {n:v for n,v in zip(centrality.keys(), assign_colors(centrality.values(), cmap=plt.cm.Greys))}, 'color')

net = Network(notebook=True, height='500px', width='100%', bgcolor="white")
net.from_nx(G)
net.show('centrality.html')
```

## Preferential attachment

[https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.link_prediction.preferential_attachment.html#networkx.algorithms.link_prediction.preferential_attachment](https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.link_prediction.preferential_attachment.html#networkx.algorithms.link_prediction.preferential_attachment)

```python
import networkx as nx
import pandas as pd

G = make_graph()
G = G.to_undirected()

neighbors = list(nx.neighbors(G, "Panizo")) + ["Panizo"]
# n es cualquier nodo,
scores = nx.preferential_attachment(G, ebunch=[('Panizo',n) for n in G.nodes if not n in neighbors])

df = pd.DataFrame(scores, columns=["source","target","score"]).sort_values("score", ascending=False)

display(df.head(5))

nx.set_node_attributes(G, {r.target:f"{r.score}" for _,r in df.iterrows()}, 'title')
nx.set_node_attributes(G, {n:v for n,v in zip(df.target, assign_colors(df.score))}, 'color')
nx.set_node_attributes(G, {n:f"{n}" for n in G.nodes}, 'label')
nx.set_node_attributes(G, "dot", 'shape')

net = Network(notebook=True, height='500px', width='100%', bgcolor="#DCDCDC")
net.from_nx(G)
net.show('centrality.html')
```
