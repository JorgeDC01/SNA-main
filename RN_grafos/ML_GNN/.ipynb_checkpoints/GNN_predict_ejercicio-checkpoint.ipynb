```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch_geometric.datasets import MoleculeNet
from torch_geometric.nn import GATv2Conv, global_add_pool, global_mean_pool, global_max_pool
from torch_geometric.data import DataLoader
from sklearn.model_selection import train_test_split
import numpy as np

# 1. Cargar el dataset MoleculeNet HIV
dataset = MoleculeNet(root='data/MoleculeNet', name='HIV')

# Dividir los índices de las moléculas en entrenamiento y prueba
indices = np.arange(len(dataset))
train_idx, test_idx = train_test_split(indices, test_size=0.2)

# Crear las máscaras de entrenamiento y prueba
train_mask = np.zeros(len(dataset), dtype=bool)
train_mask[train_idx] = True

test_mask = np.zeros(len(dataset), dtype=bool)
test_mask[test_idx] = True

# Asignar las máscaras de entrenamiento y prueba a los datos
for data in dataset:
    data.train_mask = torch.tensor(train_mask)
    data.test_mask = torch.tensor(test_mask)
```

### Red neuronal

```python
import torch
import torch.nn as nn
from torch_geometric.nn import GATConv, global_mean_pool
from torch_geometric.nn import MLP

class GAT_JumpingKnowledge_MLP(nn.Module):
    def __init__(self, node_features, hidden_channels, num_classes, num_heads=8):
        super(GAT_JumpingKnowledge_MLP, self).__init__()

        # Definimos las tres capas GAT (estándar, no v2)
        self.conv1 = GATConv(node_features, hidden_channels, heads=num_heads)
        self.conv2 = GATConv(hidden_channels * num_heads, hidden_channels, heads=num_heads)
        self.conv3 = GATConv(hidden_channels * num_heads, hidden_channels, heads=num_heads)

        # Jumping Knowledge: Concatenar las salidas de las tres capas
        self.jump_concat = True  # Usamos la concatenación

        # Capa de pooling global (puede ser global_mean_pool, global_max_pool, etc.)
        self.pool = global_mean_pool  # Usamos global_mean_pool, pero puedes cambiarlo si lo necesitas

        # Cambiamos el MLP para que acepte las características de salida correctamente
        self.cls = MLP(
            in_channels=hidden_channels * num_heads * 3,  # Ajustar a las características concatenadas
            out_channels=num_classes,
            num_layers=2,
            hidden_channels=hidden_channels,
            act='relu',
            dropout=0.5
        )

    def forward(self, data):
        # Extraemos los datos de entrada
        x, edge_index, batch = data.x, data.edge_index, data.batch

        # Aseguramos que las características del nodo sean float y los índices de las aristas sean long
        x = x.float()
        edge_index = edge_index.long()

        # Aplicamos las tres capas de GAT (estándar)
        out1 = self.conv1(x, edge_index)
        out2 = self.conv2(out1, edge_index)
        out3 = self.conv3(out2, edge_index)

        # Jumping Knowledge: concatenamos las salidas de cada capa
        if self.jump_concat:
            out = torch.cat([out1, out2, out3], dim=1)  # Concatenar las características

        # Aplicamos el pooling global
        out = self.pool(out, batch)

        # Clasificación con el MLP
        out = self.cls(out)

        return out
```

### Entrenamiento, test

```python
import torch
import torch.optim as optim
import torch.nn as nn
from torch_geometric.data import DataLoader

def train(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0
    for data in loader:
        optimizer.zero_grad()
        out = model(data)
        
        # Asegúrate de que las etiquetas están en la forma correcta
        if data.y.dim() > 1:  # Si son one-hot, conviértelas a índices de clase
            data.y = data.y.argmax(dim=1)
        
        loss = criterion(out, data.y)  # Calcula la pérdida
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

def evaluate(model, loader):
    model.eval()
    correct = 0
    total = 0
    for data in loader:
        out = model(data)
        pred = out.argmax(dim=1)  # Predicción para cada nodo
        correct += pred.eq(data.y).sum().item()  # Predicciones correctas
        total += data.y.size(0)  # Total de nodos en el lote
    return correct / total  # Exactitud

def run_experiment(pooling_type):
    # Crear el modelo
    model = GAT_JumpingKnowledge_MLP(node_features=dataset.num_node_features, 
                                      hidden_channels=64, 
                                      num_classes=2)  # Asegúrate de que num_classes es 2 o lo que corresponda

    # Definir optimizador y criterio
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    criterion = nn.CrossEntropyLoss()

    # Crear DataLoaders para entrenamiento y prueba
    train_loader = DataLoader([dataset[i] for i in train_idx], batch_size=32, shuffle=True)
    test_loader = DataLoader([dataset[i] for i in test_idx], batch_size=32, shuffle=False)

    # Entrenar el modelo
    for epoch in range(1, 61):
        loss = train(model, train_loader, optimizer, criterion)
        if epoch % 5 == 0:
            print(f"Epoch {epoch}: Loss {loss:.4f}")
    
    # Evaluar el modelo
    accuracy = evaluate(model, test_loader)
    print(f'{pooling_type} pooling test accuracy: {accuracy:.4f}')
    return accuracy
```

### main

```python
# 5. Probar los diferentes operadores de pooling
pooling_operators = ['mean', 'add', 'max']
results = {}

for pooling_type in pooling_operators:
    # Asignar el pooling correspondiente
    if pooling_type == 'mean':
        GAT_JumpingKnowledge_MLP.pool = global_mean_pool
    elif pooling_type == 'add':
        GAT_JumpingKnowledge_MLP.pool = global_add_pool
    elif pooling_type == 'max':
        GAT_JumpingKnowledge_MLP.pool = global_max_pool
    
    accuracy = run_experiment(pooling_type)
    results[pooling_type] = accuracy

# Mostrar los resultados
print("\nResults of different pooling types:")
for pooling_type, accuracy in results.items():
    print(f"{pooling_type} pooling: {accuracy:.4f}")
```

### Cargar datos a pytorch_geometric

```python
import torch
from torch_geometric.utils.convert import from_networkx
from sentence_transformers import SentenceTransformer

# transformar descripciones en embeddings de texto
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
descriptions_embeddings = model.encode(list(node_descriptions.values()))
nx.set_node_attributes(G, {node: torch.tensor(embedding) for node, embedding in zip(node_descriptions.keys(), descriptions_embeddings)}, 'x')

# Crear el grafo en PyG
data = from_networkx(nx.relabel_nodes(G, node_map))

print(f"Attributes: {data.keys()}")
print(f"Number of nodes: {data.num_nodes}")
print(f"Number of edges: {data.num_edges}")
print(f"Number of node features: {data.num_node_features}")
print(f"Number of edge features: {data.num_edge_features}")
print(f"Contains isolated nodes: {data.has_isolated_nodes()}")
print(f"Contains self-loops: {data.has_self_loops()}")
print(f"Is undirected: {data.is_undirected()}")
```

## Defino GNN

Usaremos el modulo de pytorch que implementa una [GCN](https://pytorch-geometric.readthedocs.io/en/2.5.0/generated/torch_geometric.nn.models.GCN.html)

```python
from torch_geometric.nn import GCN

gnn = GCN(
    in_channels=data.num_node_features,
    out_channels=32,
    hidden_channels=32,
    num_layers=4
)
```

## Optimizar

```python
from torch.nn.functional import sigmoid
from torch.optim import Adam

gnn.reset_parameters()

def loss_func(data, A, k=5):
    Z = gnn(data.x, data.edge_index) # calculo embeddings
    A_hat = sigmoid(Z @ Z.T) #aplico sigmoid de la funcion
    # Evito que el valor sea 0 o 1 exactamente para evitar problemas con el logaritmo
    A_hat = torch.clamp(A_hat, min=1e-7, max=1 - 1e-7) # 
    loss = -torch.sum(k*(A * torch.log(A_hat)) + (1 - A) * torch.log(1 - A_hat))
    return loss

A = torch.tensor(nx.to_numpy_array(G))

optimizer = Adam(gnn.parameters(), lr=0.005, weight_decay=5e-4)

gnn.train()
for epoch in range(2001):
    optimizer.zero_grad()
    loss = loss_func(data, A, 10)
    loss.backward()
    optimizer.step()
    
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Loss {loss.item()}")
```
